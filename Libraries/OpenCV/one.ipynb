```python
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('image1.jpg')
print(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
print(gray.shape)
```

```python
resized = cv2.resize(img, (128, 128))
```

```python
edges1 = cv2.Canny(gray, 50, 100)
edges2 = cv2.Canny(gray, 100, 200)
edges3 = cv2.Canny(gray, 150, 250)
```

```python
plt.figure(figsize=(25,25))
plt.subplot(1,3,1)
plt.imshow(edges1,cmap='gray')
plt.subplot(1,3,2)
plt.imshow(edges2,cmap='gray')
plt.subplot(1,3,3)
plt.imshow(edges3,cmap='gray')
```

Changing brightness and contrast

```python
imgn = cv2.imread('image2.png')
```

```python
plt.imshow(imgn)
```

```python
image2 = cv2.addWeighted(imgn, 1.5, np.zeros(imgn.shape, imgn.dtype), 0, 10)
# cv2.imwrite('modified_image.jpg', image2)
plt.imshow(image2)
# plt.show()
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
```

```python
filtered_image2 = cv2.GaussianBlur(imgn, (17, 17), 0)
# plt.imshow(filtered_image2)
plt.imshow(cv2.cvtColor(filtered_image2, cv2.COLOR_BGR2RGB))
plt.imshow(cv2.cvtColor(filtered_image2, cv2.COLOR_BGR2GRAY))
```
