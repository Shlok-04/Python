```python
img1 = cv.imread('image2.png')
cv.imshow('img1',img1)
print(img1.shape)
img1 = cv.resize(img1 , (800,700))  # (width , height)
print(img1.shape)
cv.imshow('resized',img1)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
M = np.float32([[1, 0, 100], [0, 1, 30]])
translated = cv.warpAffine(img1, M, (img1.shape[1], img1.shape[0])) # (width , height)
cv.imshow("Translated Image", translated)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
h = img1.shape[0]
w = img1.shape[1]
center = (w//2,h//2)
M = cv.getRotationMatrix2D(center,-45,1)
rotated = cv.warpAffine(img1,M, (w,h))
cv.imshow('Org',img1)
cv.imshow('Rotated',rotated)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
horizontal_flip = cv.flip(img1,1)
vertical_flip = cv.flip(img1,0)
cv.imshow('org',img1)
cv.imshow('horizontal',horizontal_flip)
cv.imshow('vertical',vertical_flip)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
img1 = cv.imread('image2.png')
img1 = cv.resize(img1 , (800,700))
cv.line(img1, (500, 200), (300, 200), (0, 255, 0), 2)
cv.imshow('line',img1)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
img1 = cv.imread('image2.png')
img1 = cv.resize(img1 , (800,700))
cv.rectangle(img1,(100,10),(700,600) , (255,0,0),3)
cv.imshow('rectange',img1)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
img1 = cv.imread('image2.png')
img1 = cv.resize(img1 , (800,700))
cv.circle(img1,(400,300),250,(0,255,0),thickness=3)
cv.imshow('circle',img1)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
img1 = cv.imread('image2.png')
img1 = cv.resize(img1 , (800,700))
cv.putText(img1,"CAT",(20,50),cv.FONT_HERSHEY_COMPLEX,2,(255,255,0),thickness=3)
cv.imshow('text',img1)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
img1 = cv.imread('image2.png')
img1 = cv.resize(img1 , (800,700))
blur = cv.blur(img1, (10, 10))
cv.imshow("Org Img", img1)
cv.imshow("Averaging Blur", blur)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
blur = cv.GaussianBlur(img1,(5,5),0)
cv.imshow("Org Img", img1)
cv.imshow("Gaussian Blur", blur)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
blur = cv.medianBlur(img1, 5)
cv.imshow("Median Blur", blur)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
blur = cv.bilateralFilter(img1, 9, 75, 75)
cv.imshow("Bilateral Filter", blur)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
sobelx = cv.Sobel(img1, cv.CV_64F, 1, 0, ksize=1)
sobely = cv.Sobel(img1, cv.CV_64F, 0, 1, ksize=1)

cv.imshow("Sobel X", sobelx)
cv.imshow("Sobel Y", sobely)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
sobelx1 = cv.Sobel(img1, cv.CV_64F, 1, 0, ksize=1)
sobelx2 = cv.Sobel(img1, cv.CV_64F, 1, 0, ksize=15)

cv.imshow("Sobel X1", sobelx1)
cv.imshow("Sobel X2", sobelx2)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
laplacian = cv.Laplacian(img1, cv.CV_64F)
cv.imshow("Laplacian", laplacian)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
canny = cv.Canny(img1,50,100)
cv.imshow('Canny',canny)
cv.waitKey(0)
cv.destroyAllWindows()
```
