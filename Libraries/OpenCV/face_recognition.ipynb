```python
import cv2 as cv
import pandas as pd
import numpy as np
```

```python
img1 = cv.imread('face_image_2.jpeg')
img1 = cv.resize(img1,(600,500))
cv.imshow('face image', img1)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
gray_img = cv.cvtColor(img1,cv.COLOR_BGR2GRAY)
```

```python
face_cascade = cv.CascadeClassifier("haar_face.xml")
```

```python
faces = face_cascade.detectMultiScale(gray_img,scaleFactor=1.1,minNeighbors=2)
```

```python
for (x,y,w,h) in faces:
    cv.rectangle(img1,(x,y),(x+w,y+h),(0,255,0),thickness=2)
```

```python
cv.imshow('gray image',img1)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
img2 = cv.imread('face_image_2.jpeg')
img2 = cv.resize(img2,(500,500))
gray_img = cv.cvtColor(img2,cv.COLOR_RGB2GRAY)
haar_face = cv.CascadeClassifier('haar_face.xml')
face_dect = haar_face.detectMultiScale(img2,scaleFactor=1.1,minNeighbors=2)
for (x,y,w,h) in face_dect:
    cv.rectangle(img2,(x,y),(x+w,y+h),(255,0,0),thickness=2)
cv.imshow('image of cat',img2)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
img = cv.imread('face_image_1.png')
img = cv.resize(img,(800,600))
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
haar_face = cv.CascadeClassifier('haar_face.xml')
face_dect = haar_face.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=3)
for (x,y,w,h) in face_dect:
    cv.rectangle(img,(x,y),(x+w,y+h),(200,200,0),thickness=2)
cv.imshow('image',img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
img = cv.imread('face_image_3.jpeg')
img = cv.resize(img,(500,500))
haar_face = cv.CascadeClassifier('haar_face.xml')
face_dect = haar_face.detectMultiScale(img,scaleFactor=1.1,minNeighbors=3)
count = 0
for (x,y,w,h) in face_dect:
    cv.rectangle(img,(x,y),(x+w,y+h),(255,0,0),thickness=2)
    count = count+1
cv.putText(img,f'Total count : {count}',(50,50),cv.FONT_HERSHEY_SIMPLEX,1.5,(0,0,255),thickness=2)
cv.imshow('image',img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
cap = cv.VideoCapture(0)
haar_face = cv.CascadeClassifier('haar_face.xml')
while True:
    ret, frame = cap.read()
    face_dect = haar_face.detectMultiScale(frame,scaleFactor=1.1,minNeighbors=1)
    count = 0
    for (x,y,w,h) in face_dect:
        cv.rectangle(frame,(x,y),(x+w,y+h), (0,255,0),thickness=2)
        count+=1
    cv.putText(frame,f'Total count : {count}',(50,50),cv.FONT_HERSHEY_SIMPLEX,1.5,(0,0,255),thickness=2)
    cv.imshow('face',frame)
    if cv.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv.destroyAllWindows()
```

```python
cap = cv.VideoCapture(0)
ret, frame1 = cap.read()
ret, frame2 = cap.read()

while True:
    diff = cv.absdiff(frame1, frame2)
    gray = cv.cvtColor(diff, cv.COLOR_BGR2GRAY)
    blur = cv.GaussianBlur(gray, (5, 5), 0)
    _, thresh = cv.threshold(blur, 20, 255, cv.THRESH_BINARY)

    cv.imshow("Motion", thresh)

    frame1 = frame2
    ret, frame2 = cap.read()
    if cv.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```

```python
fgbg = cv.createBackgroundSubtractorMOG2()
cap = cv.VideoCapture(0)
while True:
    ret, frame = cap.read()
    fgmask = fgbg.apply(frame)

    cv.imshow("Foreground Mask", fgmask)

    if cv.waitKey(1) & 0xFF == ord('q'):
        break
```
