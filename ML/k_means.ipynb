```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.cluster import KMeans
```

```python
df = pd.DataFrame({
    'annual_income': [15, 16, 17, 25, 26, 27, 60, 62, 63, 64],
    'spending_score': [39, 81, 6, 77, 40, 6, 50, 49, 48, 52]
})

print(df)
```

```python
plt.scatter(df['annual_income'],df['spending_score'])
plt.xlabel("annual_income")
plt.ylabel("spending_score")
```

```python
x = df.iloc[:,:]
x
```

```python
km = KMeans(n_clusters=3,random_state=42)
km.fit(x)
```

```python
df['cluster'] = km.labels_
```

```python
df
```

```python
plt.scatter(df['annual_income'],df['spending_score'],c=df['cluster'])
```

```python
km = KMeans(n_clusters=4,random_state=42)
km.fit(x)
```

```python
df['cluster1'] = km.labels_
print(df)
```

```python
plt.scatter(df['annual_income'],df['spending_score'],c=df['cluster1'])
```
