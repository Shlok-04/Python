```python
df = pd.DataFrame({
    'age': [20, None, 30],
    'gender': ['M', 'F', 'M'],
    'income': [40000, 50000, None]
})
df
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
df_imp = imputer.fit_transform(df.drop(columns=['gender']))
df[['age','income']] = df_imp
print(df)
```
