```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'text': [
        'Congratulations you have won a lottery',
        'You won a free ticket',
        'Please call me when you reach',
        'Let us meet tomorrow',
        'Claim your free prize now',
        'Are you coming to office today?'
    ],
    'label': ['spam', 'spam', 'ham', 'ham', 'spam', 'ham']
})

print(df)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
enc_label = le.fit_transform(df['label'])
df['encoded_label'] = enc_label
print(df)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
x = cv.fit_transform(df['text'])
y = df['encoded_label']
```

```python
x.toarray()
```

```python
x.toarray(),y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.1,random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(xtrain, ytrain)
```

```python
df
```

```python
ytest
```

```python
ypred = model.predict(xtest)
print("Y test\n",ytest)
print()
print("Y pred",ypred)
```

```python
test_email = ['Win a free iPhone now']

X_new = cv.transform(test_email)  
prediction = model.predict(X_new)

# print(le.transform(['spam']))
# print(le.inverse_transform([0]))

print("Prediction (0=ham, 1=spam):", prediction)
print("Label:", le.inverse_transform(prediction))
```
