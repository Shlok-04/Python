```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'income': [30000, 50000, 40000, 80000, 60000, 75000],
    'credit_score': [600, 700, 650, 720, 680, 710],
    'loan_amount': [15000, 20000, 12000, 30000, 25000, 28000],
    'approved': [0, 1, 0, 1, 1, 1]
})

print(df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2)
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(xtrain,ytrain)
```

```python
dtc.predict(xtest)
```

```python
ytest
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(5, 7))
plot_tree(dtc, filled=True, feature_names=['income','credit_score','loan_amount'], class_names=['Not Approved','Approved'])
plt.show()
```
