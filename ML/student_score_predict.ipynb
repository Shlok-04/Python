```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'study_hours': [2, 3, 4, 5, 6, 7],
    'sleep_hours': [7, 6, 8, 5, 7, 6],
    'practice_tests': [1, 2, 2, 3, 3, 4],
    'score': [50, 55, 65, 70, 78, 85]
})

print(df)
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2)
```

```python
lr = LinearRegression()
lr.fit(xtrain,ytrain)
```

```python
ypred = lr.predict(xtest)
print(ypred)
print(ytest)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['study_hours'], df['score'])
plt.plot(df['study_hours'], lr.predict(x), linestyle='dashed')
plt.xlabel("Study Hours")
plt.ylabel("Score")
plt.title("Study Hours vs Score Prediction")
plt.show()
```

```python
print(lr.predict([[6, 7, 3]]))
```

```python
from sklearn.metrics import r2_score
```

```python
r2_sc = r2_score(ytest,ypred)
print(r2_sc)
```
